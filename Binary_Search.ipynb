```python
#search Insert

def searchInsertK(self, Arr, N, k):
        low=0
        high=N-1
        ans=N
        while low<=high:
            mid=(low+high)//2
            if Arr[mid]>=k:
                ans=mid
                high=mid-1
            else:
                low=mid+1
        return ans
```

```python
#Floor in a Sorted Array
def findFloor(self,A,N,X):
        low=0
        high=N-1
        ans=-1
        while low<=high:
            mid=(low+high)//2
            if A[mid]<=X:
                ans=mid
                low=mid+1
                
            else:
                high=mid-1
        return ans
```

```python
#Ceil The Floor

def getFloorAndCeil(self, x: int, arr: list) -> list:
            floor = -1
            ceil = -1
        
            for num in arr:
                if num <= x:
                    if num > floor:
                        floor = num
                if num >= x:
                    if ceil == -1 or num < ceil:
                        ceil = num
                        
            return [floor, ceil]
```

```python
#Minimum Number in a sorted rotated array

import sys
class Solution:
    
    #Function to find the minimum element in sorted and rotated array.
    def minNumber(self, arr,low,high):
        ans=sys.maxsize
        while low<=high:
            mid=(low+high)//2
            
            if arr[low]<=arr[high]:
                ans=min(ans,arr[low])
            if arr[low]<=arr[mid]:
                ans=min(ans,arr[low])
                low=mid+1
            else:
                ans=min(ans,arr[mid])
                high=mid-1
        return ans
```

```python

#Find the element that appears once in sorted array
def findOnce(self,arr : list, n : int):
        if n==1:
            return arr[0]
        if arr[0]!=arr[1]:
            return arr[0]
        if arr[n-1]!=arr[n-2]:
            return arr[n-1]
        low=1
        high=n-2
        while low<=high:
            mid=(low+high)//2
             # If arr[mid] is the single element:
            if arr[mid] != arr[mid + 1] and arr[mid] != arr[mid - 1]:
                return arr[mid]
             # We are in the left:
            if (mid % 2 == 1 and arr[mid] == arr[mid - 1]) or (mid % 2 == 0 and arr[mid] == arr[mid + 1]):
                # Eliminate the left half:
                low = mid + 1
            # We are in the right:
            else:
                # Eliminate the right half:
                high = mid - 1
    
        # Dummy return statement:
        return -1
```

```python
#Peak element

def peakElement(self,arr):
        n=len(arr)
        if n==1:
            return 0
        if arr[0]>=arr[1]:
            return 0
        if arr[n-1]>=arr[n-2]:
            return n-1
        low=1
        high=n-2
        while low<=high:
            mid=(low+high)//2
            if arr[mid-1]<=arr[mid]and arr[mid]>=arr[mid+1]:
                return mid
            if arr[mid]>arr[mid-1]:
                low=mid+1
            else:
                high=mid-1
        return -1
```
