```python
#Floor in a Sorted Array
def findFloor(self,A,N,X):
        low=0
        high=N-1
        ans=-1
        while low<=high:
            mid=(low+high)//2
            if A[mid]<=X:
                ans=mid
                low=mid+1
                
            else:
                high=mid-1
        return ans
```

```python
#Ceil The Floor

def getFloorAndCeil(self, x: int, arr: list) -> list:
            floor = -1
            ceil = -1
        
            for num in arr:
                if num <= x:
                    if num > floor:
                        floor = num
                if num >= x:
                    if ceil == -1 or num < ceil:
                        ceil = num
                        
            return [floor, ceil]
```

```python
# Find the Nth root of M
def func(self,mid,n,m):
        ans=1
        for i in range(1,n+1):
            ans*=mid
            if ans>m:
                return 2
        if ans==m:
            return 1
        return 0
            
def NthRoot(self, n, m):
	    low=1
	    high=m
	    while low<=high:
	        mid=(low+high)//2
	        midN=self.func(mid,n,m)
	        if midN == 1:
                return mid
            elif midN == 0:
                low = mid + 1
            else:
                high = mid - 1
        return -1
		
```

```python
#Minimum days to make M bouquets

def possible(self,arr,days,M,K):
        n=len(arr)
        count=0
        no_of_bouquet=0
        for i in range(n):
            if arr[i]<=days:
                count+=1
            else:
                no_of_bouquet+=count//K
                count=0
        no_of_bouquet+=count//K
        return no_of_bouquet>=M

                
def solve(self, M, K, bloomDay):
        n=len(bloomDay)
        val=M*K
        if val>n:
            return -1
        low=min(bloomDay)
        high=max(bloomDay)
        while low<=high:
            mid=(low+high)//2
            ans=self.possible(bloomDay,mid,M,K)
            if ans:
                high=mid-1
            else:
                low=mid+1
            
        return low
    
```

```python
# Smallest Divisor

import math
class Solution:
    def total(self,nums,divisor):
        n=len(nums)
        total_nums=0
        for i in range(n):
            total_nums+=(nums[i] + divisor - 1) // divisor
        return total_nums
    def smallestDivisor(self, nums, K):
        low=1
        high=max(nums)
        
        while low<high:
            mid=(low+high)//2
            res=self.total(nums,mid)
            if res<=K:
                high=mid
            else:
                low=mid+1
        return low
```

```python
#Capacity To Ship Packages Within D Days

def finddays(self, arr, cap):
            n = len(arr)
            day = 1
            load = 0
            for i in range(n):
                if load + arr[i] > cap:
                    day += 1
                    load = arr[i]
                else:
                    load += arr[i]
            return day

def leastWeightCapacity(self, arr, n, d):
            low = max(arr)
            high = sum(arr)
            
            while low < high:
                mid = (low + high) // 2
                res = self.finddays(arr, mid)
                
                if res <= d:
                    high = mid  # Check for smaller capacities
                else:
                    low = mid + 1  # Increase capacity since we need more days
            
            return low
```

```python
def checkDuplicatesWithinK(self, arr, k):
        if k == 0:
            return False 
    
        # Dictionary to store the last index of each element
        seen = {}
        
        for i in range(len(arr)):
            # If the element is already in the dictionary and the difference is <= k
            if arr[i] in seen and i - seen[arr[i]] <= k:
                return True
            
            # Update the last seen index of the element
            seen[arr[i]] = i
        
        return False
```
